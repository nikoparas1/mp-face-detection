```python
import os
import glob
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
train_dir = "images/train"
annotated_dir = "images/annotated"
os.makedirs(annotated_dir, exist_ok=True)
```

```python
def show_image(img, cmap=None):
    """Display an OpenCV image (BGR) in Jupyter via matplotlib (RGB)."""
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 6))
    plt.imshow(img_rgb, cmap=cmap)
    plt.axis("off")
```

```python
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

face_detector = mp_face_detection.FaceDetection(
    model_selection=1, min_detection_confidence=0.5
)
```

```python
for img_path in glob.glob(os.path.join(train_dir, '*.jpg')):
    img = cv2.imread(img_path)
    if img is None:
        print(f"⚠️  Could not read {img_path}, skipping.")
        continue

    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = face_detector.process(rgb)

    annotated = img.copy()
    num_faces = len(results.detections) if results.detections else 0
    for det in results.detections or []:
        mp_drawing.draw_detection(annotated, det)

    out_path = os.path.join(annotated_dir, os.path.basename(img_path))
    cv2.imwrite(out_path, annotated)
    print(
        f"✅  {os.path.basename(img_path)} → {num_faces} face(s), saved to {out_path}"
    )

face_detector.close()
```
